```python
from numpy import array

i_hat1 = array([0, 1])
j_hat1 = array([-1, 0])

transform1 = array([i_hat1, j_hat1]).transpose()

print(i_hat1)
print(j_hat1)
print(transform1)
```

```python
i_hat2 = array([1, 0])
j_hat2 = array([1, 1])

transform2 = array([i_hat2, j_hat2]).transpose()

print(i_hat2)
print(j_hat2)
print(transform2)
```

```python
combined = transform1 @ transform2
print(combined)
```

```python
from numpy.linalg import det

i_hat = array([3, 0])
j_hat = array([0, 2])

basis = array([i_hat, j_hat]).transpose()

determinant = det(basis)
print(determinant)
```

```python
from sympy import *
A = Matrix([
    [4,2,4],
    [5,3,7],
    [9,3,6]
])

inverse = A.inv()
identity = inverse*A

print('역행렬 :', inverse)
print('항등행렬 :', identity)
```

```python
from numpy import diag, array
from numpy.linalg import eig, inv

A = array([
    [1,2],
    [4,5]
])

eigenvals, eigenvecs = eig(A)

print(eigenvals)
print(eigenvecs)
```
