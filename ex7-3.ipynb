```python
all_inputs = (all_data.iloc[:, 0:3].values / 255.0)
```

```python
all_outputs = all_data.iloc[:,-1].values
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(all_inputs, all_outputs, test_size = 1/3)
```

```python
w_hidden = np.random.rand(3,3)
```

```python
w_hidden
```

```python
w_output = np.random.rand(1,3)
```

```python
w_output
```

```python
b_hidden = np.random.rand(3, 1)
```

```python
b_output = np.random.rand(1, 1)
```

```python
relu = lambda x: np.maximum(x, 0)
logistic = lambda x: 1 / (1 + np.exp(-x))
```

```python
def forward_prop(X) :
    Z1 = w_hidden @ X + b_hidden
    A1 = relu(Z1)
    Z2 = w_output @ A1 + b_output
    A2 = logistic(Z2)
    return Z1, A1, Z2, A2
```

```python
test_predictions = forward_prop(X_train.transpose())[3]
```

```python
test_comparisons = np.equal((test_predictions >= 0.5).flatten().astype(int), Y_train)
```

```python
accuracy = sum(test_comparisons.astype(int) / X_train.shape[0])
```
